# 01 — Ingest: MSSQL → Bronze Parquet

Reads tables from SQL Server and writes them as Parquet in the **bronze** zone.

```python
import sys, os
sys.path.insert(0, '/workspace')

from scripts.spark_session import get_spark, JDBC_URL, JDBC_PROPS, BRONZE

spark = get_spark('01_ingest')
print(f'Spark version: {spark.version}')
```

```python
df_customers = spark.read.jdbc(JDBC_URL, 'dbo.customers', properties=JDBC_PROPS)
df_customers.show()
df_customers.write.mode('overwrite').parquet(f'{BRONZE}/customers')
print(f'✅ Wrote {df_customers.count()} customers to bronze')
```

```python
df_orders = spark.read.jdbc(JDBC_URL, 'dbo.orders', properties=JDBC_PROPS)
df_orders.show()
df_orders.write.mode('overwrite').parquet(f'{BRONZE}/orders')
print(f'✅ Wrote {df_orders.count()} orders to bronze')
```
